```python
import os
import zipfile
```

```python

# Ścieżka bazowa
base_dir = os.path.normpath(r"data/larfield_sample/iteration_04")

# Wspólny katalog docelowy
extract_dir = os.path.normpath(os.path.join(base_dir, "extracted_files"))
os.makedirs(extract_dir, exist_ok=True)

# Iteracja po wszystkich plikach w katalogu bazowym
for root, _, files in os.walk(base_dir):
    for file in files:
        if file.endswith(".zip"):
            zip_path = os.path.normpath(os.path.join(root, file))
            print(f"Rozpakowywanie: {zip_path}")

            # Wyznaczenie relatywnej ścieżki względem katalogu bazowego
            relative_path = os.path.relpath(root, base_dir)
            relative_path = os.path.normpath(os.path.join(*relative_path.split(os.sep)[1:]))
            target_dir = os.path.normpath(os.path.join(extract_dir, relative_path))

            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                for member in zip_ref.namelist():
                    if member.endswith('/'):
                        continue
                    # Tworzenie pełnej ścieżki docelowej
                    target_path = os.path.normpath(os.path.join(target_dir, member))
                    os.makedirs(os.path.dirname(target_path), exist_ok=True)
                    with zip_ref.open(member) as source, open(target_path, 'wb') as target:
                        target.write(source.read())
                print(f" → Rozpakowano do: {target_path}")
```
