```python
import pandas as pd
import neurokit2 as nk
```

```python
# Wczytanie danych z pliku ecg.csv
ecg_data = pd.read_csv(
    'data/larfield_sample/iteration_04/extracted_files/2VaupA1XhdeElSqKXSxzGQljhT02/POLAR_B2E7B722_v1_2023-06-26T10-07-00-664/ECG.csv',
    sep='\t'
)
print(ecg_data.head())
ecg_signal = ecg_data['ecg']

# Przetwarzanie sygnału ECG
ecg_cleaned = nk.ecg_clean(ecg_signal, sampling_rate=1000)  # Dopasuj sampling_rate do danych

# Ocena jakości sygnału ECG
quality = nk.ecg_quality(ecg_cleaned, sampling_rate=1000)
print(f"Jakość sygnału: {quality.mean()}")

# Kryteria oceny jakości
if quality.mean() > 0.5:  # Próg można dostosować
    print("Sygnał jest dobrej jakości.")
else:
    print("Sygnał jest złej jakości.")
```

```python
import numpy as np
import os
import pandas as pd
import neurokit2 as nk
import json

base_dir = r"data/larfield_sample/iteration_04/extracted_files"

for folder in os.listdir(base_dir):
    folder_path = os.path.join(base_dir, folder)
    print(f"Przetwarzanie folderu: {folder_path}")
    if os.path.isdir(folder_path):
        metrics_dict = {}
        for subfolder in os.listdir(folder_path):
            print(f"Przetwarzanie podfolderu: {subfolder}")
            subfolder_path = os.path.join(folder_path, subfolder)
            if os.path.isdir(subfolder_path):
                print(f"Znaleziono podfolder: {subfolder_path}")
                ecg_file = os.path.join(subfolder_path, "ECG.csv")
                if os.path.exists(ecg_file) and os.path.getsize(ecg_file) > 0:
                    print(f"Znaleziono plik ECG: {ecg_file}")
                    try:
                        ecg_data = pd.read_csv(ecg_file, sep='\t')
                    except pd.errors.EmptyDataError:
                        continue
                    if ecg_data.empty or 'ecg' not in ecg_data.columns:
                        print(f"Plik ECG jest pusty lub nie zawiera kolumny 'ecg': {ecg_file}")
                        continue
                    ecg_signal = ecg_data['ecg']
                    quality = [0.8, 0.6, 0.7]  # Przykładowe wartości jakości, zastąp rzeczywistymi obliczeniami
                    mean_quality = float(np.mean(quality))
                    metrics = {
                        "mean_quality": mean_quality,
                        "is_good_quality": bool(mean_quality > 0.5)
                    }
                    metrics_dict[subfolder] = metrics
        metrics_file = os.path.join(folder_path, "metrics.json")
        with open(metrics_file, "w", encoding="utf-8") as f:
            json.dump(metrics_dict, f, ensure_ascii=False, indent=4)
```
